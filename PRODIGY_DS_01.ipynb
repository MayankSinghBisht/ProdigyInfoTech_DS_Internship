```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Population.csv",skiprows=4)

data.head()
```

```python
population_2020 = data[["Country Name", "2020"]].dropna()
population_2020.head()
```

```python
plt.figure()
plt.hist(population_2020["2020"], bins=20)
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.title("Distribution of Population by Country (2020)")
plt.show()
```

```python
top_10 = population_2020.sort_values(by="2020", ascending=False).head(10)

plt.figure()
plt.bar(top_10["Country Name"], top_10["2020"])
plt.xticks(rotation=45, ha="right")
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Top 10 Most Populated Countries (2020)")
plt.show()
```
