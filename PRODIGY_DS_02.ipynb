```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Titanic.csv')
```

```python
print(data.head())
print(data.info())
print(data.isnull().sum())
```

```python
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
data['Fare'].fillna(data['Fare'].median(), inplace=True)
data.drop('Cabin', axis=1, inplace=True) 
data.drop('PassengerId', axis=1, inplace=True)
```

```python
sns.countplot(x='Survived', data=data)
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=data)
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=data)
plt.show()
```

```python
sns.histplot(data=data, x='Age', hue='Survived', multiple='stack')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Fare', data=data)
plt.show()
```

```python
data['FamilySize'] = data['SibSp'] + data['Parch'] + 1
sns.countplot(x='FamilySize', hue='Survived', data=data)
plt.show()
```

```python
print("Survival rate by gender:\n", data.groupby('Sex')['Survived'].mean())
print("\nSurvival rate by class:\n", data.groupby('Pclass')['Survived'].mean())
print("\nSurvival rate by family size:\n", data.groupby('FamilySize')['Survived'].mean())
```
